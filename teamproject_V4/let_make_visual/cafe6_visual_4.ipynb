```python
import seaborn as sns
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
df=pd.read_csv('acp.csv')
```

```python
df=df.replace('강남구','g1')
```

```python
df=df.replace('서초구','g2')
```

```python
df=df.replace('송파구','g3')
```

```python
df=df.replace('강북구','b1')
```

```python
df=df.replace('노원구','b2')
```

```python
df=df.replace('도봉구','b3')
```

```python
df=df.replace('성북구','b4')
```

```python
df
```

```python
g3=df[(df['gu_num']==1)]
```

```python
g3
```

```python
bed4=df[(df['gu_num']==0)]
```

```python
bed4
```

```python
sns.regplot(x="high_count", y="high_bus_sum", data=g3) #0.9
```

```python
sns.regplot(x="high_count", y="high_bus_sum", data=bed4) #0.6
```

```python
sns.regplot(x="high_count", y="high_sub_average", data=g3)
```

```python
sns.regplot(x="high_count", y="high_sub_average", data=bed4)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2,figsize=(15,15))
sns.regplot(x="high_count", y="high_bus_sum", data=g3, ax=ax[0,0])
sns.regplot(x="high_count", y="high_bus_sum", data=bed4,ax=ax[0,1])
sns.regplot(x="high_count", y="high_sub_average", data=g3,ax=ax[1,0])
sns.regplot(x="high_count", y="high_sub_average", data=bed4,ax=ax[1,1])
```

```python
fig, ax = plt.subplots(ncols=2,figsize=(20,10))
sns.regplot(x="high_count", y="low_count", data=g3, ax=ax[0]).set_title('G3',size=50)
sns.regplot(x="high_count", y="low_count", data=bed4,ax=ax[1]).set_title('D4',size=50)
```

```python
fig, ax = plt.subplots(ncols=2,figsize=(20,10))
sns.regplot(x="high_count", y="low_bus_sum", data=g3, ax=ax[0]).set_title('G3',size=50)
sns.regplot(x="high_count", y="low_bus_sum", data=bed4,ax=ax[1]).set_title('D4',size=50)
```

```python
fig, ax = plt.subplots(ncols=2,figsize=(20,10))
sns.regplot(x="low_count", y="high_bus_sum", data=g3, ax=ax[0]).set_title('G3',size=50)
sns.regplot(x="low_count", y="high_bus_sum", data=bed4,ax=ax[1]).set_title('D4',size=50)
```

```python
import json
```

```python
import pandas as pd

# 미리 만들어둔 데이터를 불러옵니다.

df = pd.read_csv('cafe6.csv',encoding='ansi')
df.head()
# import csv
# df = open('seoul_starbucks.csv','r')
# df
```

```python
geo_data = json.load(open('seoul_map.geojson', encoding='utf8'))
```

```python
# 서울시 중심부의 위도, 경도 입니다.
center = [37.541, 126.986]

# 맵이 center 에 위치하고, zoom 레벨은 11로 시작하는 맵 m을 만듭니다.
m1 = folium.Map(location=center, zoom_start=10)

# Choropleth 레이어를 만들고, 맵 m에 추가합니다.
folium.Choropleth(
    geo_data=geo_data,
    data=df,
    columns=('시군구명','star'),
    key_on='feature.properties.SIG_KOR_NM',
    fill_color='BuPu',
    legend_name='스타벅스 수',
).add_to(m1)

# 맵 m을 출력합니다.
m1
```

```python
for a in df.index:
    gu=df.loc[a,'상호명']
    if gu in("스타벅스"):
        lat = df.loc[a,'위도']
        lng = df.loc[a,'경도']
        folium.Circle([lat,lng],popup=df.loc[a,'상가업소번호'],
                      color = 'blue',
                  tooltip=df.loc[a,'시군구명']).add_to(m1)
m1
```

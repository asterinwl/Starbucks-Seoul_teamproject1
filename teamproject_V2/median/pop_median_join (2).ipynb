```python
gus =[]
for gu in pop['gu']:
    gus.append(gu)
guss=set(gus)
```

```python
gus_median = pd.DataFrame(columns=['id_popul','id_gu','gu','median','mean','year'])
```

```python
for year in set(pop['year']): 
    for gus in guss:
        gus_median=gus_median.append(pd.DataFrame([[0,0,gus,
                        round(pop[(pop['year']==year) & (pop['gu']==gus)].median()[3]),
                        round(pop[(pop['year']==year) & (pop['gu']==gus)].mean()[3]),
                          year]],columns=['id_popul','id_gu','gu','median','mean','year']),ignore_index=True)
gus_median
```

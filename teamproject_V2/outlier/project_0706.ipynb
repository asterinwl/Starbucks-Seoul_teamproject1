```python
def outliers_iqr(data):
    data1=data.iloc[:,3]
    q1, q3 = np.percentile(data1,[2,75])
    iqr = q3 - q1
    lower_bound = q1 - (iqr * 1.5)
    upper_bound = q3 + (iqr * 1.5)
    result_ind=np.where((data1 > upper_bound)| (data1 < lower_bound))
    result = delData(data,result_ind[0])
    return result
```

```python
def delData(data, index):
    for i in index:
        index = data.iloc[i,-1]
        try:
            data.drop(index = data[index], axis=0, inplace=True)
        except Exception as e:
            print(e)
    return data
```

```python
# re = pd.DataFrame(columns=['ind','gap'])
re = []
for ye in set(pop['year']):
    for i in set(pop['id']):
        a = pop[(pop['year']==ye) & (pop['id']==i)]
        result= outliers_iqr(a)
re.append(result)
```

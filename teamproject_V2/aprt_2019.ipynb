```python
import pandas as pd
```

```python
aprt_2019=pd.read_csv('./apart/aprt_2018.csv', encoding='euc-kr')
aprt_2019=pd.read_csv('./apart/aprt_2019.csv', encoding='euc-kr')
aprt_2020=pd.read_csv('./apart/aprt_2020.csv', encoding='euc-kr')
```

```python
aprt_2019.columns
```

```python
aprt_2019_median=aprt_2019.groupby([u'구'],as_index=False)[u'전용면적'].median()
aprt_2019_median.rename(columns={u'전용면적':u'전용면적_median'},inplace=True)
```

```python
aprt_2019_max=aprt_2019.groupby([u'구'],as_index=False)[u'전용면적'].max()
aprt_2019_max.rename(columns={u'전용면적':u'전용면적_max'},inplace=True)
```

```python
aprt_2019_min=aprt_2019.groupby([u'구'],as_index=False)[u'전용면적'].min()
aprt_2019_min.rename(columns={u'전용면적':u'전용면적_min'},inplace=True)
```

```python
aprt_2019_mean=aprt_2019.groupby([u'구'],as_index=False)[u'전용면적'].mean()
aprt_2019_mean.rename(columns={u'전용면적':u'전용면적_mean'},inplace=True)
```

```python
aprt_2019_1=pd.merge(aprt_2019_median,aprt_2019_max, on=None)
```

```python
aprt_2019_1=pd.merge(aprt_2019_1,aprt_2019_min, on=None)
```

```python
aprt_2019_1=pd.merge(aprt_2019_1,aprt_2019_mean, on=None)
```

```python
aprt_2019_1
```

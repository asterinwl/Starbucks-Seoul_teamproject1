```python
import pandas as pd
```

```python
aprt_2018=pd.read_csv('./apart/aprt_2018.csv', encoding='euc-kr')
aprt_2019=pd.read_csv('./apart/aprt_2019.csv', encoding='euc-kr')
aprt_2020=pd.read_csv('./apart/aprt_2020.csv', encoding='euc-kr')
```

```python
aprt_2018.columns
```

```python
aprt_2018_median=aprt_2018.groupby([u'구'],as_index=False)[u'전용면적'].median()
aprt_2018_median.rename(columns={u'전용면적':u'전용면적_median'},inplace=True)
```

```python
aprt_2018_max=aprt_2018.groupby([u'구'],as_index=False)[u'전용면적'].max()
aprt_2018_max.rename(columns={u'전용면적':u'전용면적_max'},inplace=True)
```

```python
aprt_2018_min=aprt_2018.groupby([u'구'],as_index=False)[u'전용면적'].min()
aprt_2018_min.rename(columns={u'전용면적':u'전용면적_min'},inplace=True)
```

```python
aprt_2018_mean=aprt_2018.groupby([u'구'],as_index=False)[u'전용면적'].mean()
aprt_2018_mean.rename(columns={u'전용면적':u'전용면적_mean'},inplace=True)
```

```python
aprt_2018_1=pd.merge(aprt_2018_median,aprt_2018_max, on=None)
```

```python
aprt_2018_1=pd.merge(aprt_2018_1,aprt_2018_min, on=None)
```

```python
aprt_2018_1=pd.merge(aprt_2018_1,aprt_2018_mean, on=None)
```

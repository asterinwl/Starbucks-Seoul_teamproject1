### -지하철, 버스 데이터 처리

```python
import pandas as pd
```

```python
busstop_cnt=pd.read_csv('./cafe_total_busstop_cnt.csv',encoding='ansi')
```

```python
busstop_cnt.rename(columns={'상가업소번호':'code', '상호명':'cafe_nm', '시군구코드':'gu_code', '시군구명':'gu', '법정동코드':'dong_cd', '법정동명':'dong', '지번주소':'gibun',
                               '도로명':'road', '경도':'lng','위도':'lat'},inplace=True)
```

```python
busstop_cnt
```

```python
busstop_cnt=busstop_cnt.drop(columns=['field_1'])
```

```python
busstop_cnt
```

```python
subway_dis=pd.read_csv('./cafe_total_station_distance.csv',encoding='ansi')
```

```python
subway_dis.rename(columns={'상가업소번호':'code', '상호명':'cafe_nm', '시군구코드':'gu_code', '시군구명':'gu', '법정동코드':'dong_cd', '법정동명':'dong', '지번주소':'gibun',
                               '도로명':'road', '경도':'lng','위도':'lat'},inplace=True)
```

```python
subway_dis
```

```python
subway_dis=subway_dis.drop(columns=['field_1'])
```

```python
subway_dis
```

```python
bus_subway_hap=pd.merge(busstop_cnt, subway_dis,left_on=['code','road','lng','lat','year'],right_on=['code','road','lng','lat','year'],how='inner')
#'code','road','lng','lat','year'
#bus_subway_hap=pd.merge(busstop_cnt, subway_dis,on=['code','gu_code','dong_cd', 'year'])
```

```python
bus_subway_hap
```

```python
bus_subway_hap=pd.merge(busstop_cnt, subway_dis,on=['code','cafe_nm','gu_code','gu','dong_cd','dong','gibun','road','lng','lat','year'])
```

```python
bus_subway_hap
```

```python
bus_subway_hap.to_csv('./bus_subway1.csv',encoding='ansi')
```

```python
bus_subway_hap=bus_subway_hap.drop_duplicates()  #중복된 값 없애기
```

```python
bus_subway_hap 
```

```python
bus_subway_hap.to_csv('./bus_subway2.csv',encoding='ansi')
```

```python
bus_subway_hap['cafe_cd'] = 0
```

```python
bus_subway_hap.tail(10)
```

```python
# set(bus_subway_hap['cafe_nm'])
```

```python
for i in range(len(bus_subway_hap.index)):
    st= bus_subway_hap['cafe_nm'][i].find('스타벅스')
    if st != -1:
        bus_subway_hap['cafe_cd'][i] = 1
    pal= bus_subway_hap['cafe_nm'][i].find('폴바')
    if pal != -1:
        bus_subway_hap['cafe_cd'][i] = 1
    cofB= bus_subway_hap['cafe_nm'][i].find('커피빈')
    if cofB != -1:
        bus_subway_hap['cafe_cd'][i] = 1
```

```python
etcCafe=['COFFEEBEAN영동2교사거리점',
 'PaulBassett',
 'STARBUCKSCOFFEE']
```

```python
for name in etcCafe:
    ind = bus_subway_hap[bus_subway_hap['cafe_nm']==name].index
    for i in ind:
        bus_subway_hap.iloc[i,-1] = 1
```
